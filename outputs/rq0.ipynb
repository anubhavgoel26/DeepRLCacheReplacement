```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("baselines.txt")[['Agent', 'Size', 'Misses']]
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(data=data, x='Size', y='Misses', hue='Agent', dodge=True)
plt.xlabel('Cache size')  # Customizing x-label
plt.ylabel('Miss rate (%age)')  # Customizing y-label
```
